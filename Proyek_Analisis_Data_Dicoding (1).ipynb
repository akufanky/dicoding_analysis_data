## Visualization & Explanatory Analysis

**Pertanyaan 1 : Hari apa orang - orang biasanya ramai memakai sepeda?**

```python
data_harian = hour_df.groupby('weekday')['count'].sum().reset_index()

label_harian = {
    0: 'Minggu',
    1: 'Senin',
    2: 'Selasa',
    3: 'Rabu',
    4: 'Kamis',
    5: 'Jumat',
    6: 'Sabtu'
}

data_harian['weekday'] = data_harian['weekday'].map(label_harian)

data_harian = data_harian.sort_values('weekday', key=lambda x: x.map(label_harian))

plt.figure(figsize=(10, 6))
sns.barplot(x='weekday',
            y='count',
            data=data_harian,
            palette='viridis')
plt.title('Jumlah Penyewaan Sepeda Harian')
plt.xlabel('Hari')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

**Pertanyaan 2 : Jam berapa biasanya orang - orang menggunakan sepeda?**

```python
data_perjam = hour_df.groupby('hour')['count'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='hour',
             y='count',
             data=data_perjam,
             marker='o',
             color='green')
plt.title('Jumlah Penyewaan Sepeda Per Jam')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Conclusion

- Conclusion pertanyaan 1: Dilihat dari visualisasi diatas, bisa disimpulkan bahwa: orang - orang akan banyak menyewa sepeda pada hari **Kamis** dan **Jum'at**. Kenapa ada 2 hari, yaitu Kamis dan Jum'at? Karena, selisih diantara 2 hari tersebut hanya terpaut koma(,) saja. Sehingga kalau dibulatkan akan sama hasilnya.

- Conclusion pertanyaan 2: Dan dilihat dari visualisasi waktu penyewaan sepeda, didapat bahwa pada jam **17:00** atau **5PM** orang banyak menyewa sepeda, mungkin saja karena jam tersebut adalah jam pulang sekolah maupun jam pulang kerja. Sehingga di jam tersebut banyak orang menggunakan sepeda.
